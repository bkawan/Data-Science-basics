# Data extraction
- Total Medal
- total gold medal
- total silver medal
- total bronze medal
- total medal men
    - total gold medal men
    - total silver medal men
    - total bronze medal men
- total medal women
    - total gold medal women
    - total silver medal women
    - total bronze medal women
   

```python
total_medal=[]

total_gold = []
total_silver = []
total_bronze =[]
total_medal_men = []
total_medal_women = []
total_gold_medal_men =[]
total_silver_medal_men =[]
total_bronze_medal_men =[]
total_gold_medal_women = []
total_silver_medal_women =[]
total_bronze_medal_women =[]




i=0
j =0
while i < countries_columns:
#     print(i)
    while j < len(all_country):
        #print(total_country[j])
        country = df['NOC'] == all_country[j]
#         print(total_country[j])
#         print(df[country].Medal.count())
        total_medal.append(df[country].Medal.count())
        country_df = df[country]
        men = country_df['Gender'] == 'Men'
        women = country_df['Gender'] == 'Women'
        gold = country_df['Medal'] =='Gold'
        silver = country_df['Medal'] =='Silver'
        bronze = country_df['Medal'] =='Bronze'
        
        total_medal.append(country_df.Medal.count())
        total_gold.append(country_df[gold].Medal.count())
        total_silver.append(country_df[silver].Medal.count())
        total_bronze.append(country_df[bronze].Medal.count())
        total_medal_men.append(country_df[men].Medal.count())
        total_medal_women.append(country_df[women].Medal.count())
        total_gold_medal_men.append(country_df[men][gold].Medal.count())
        total_silver_medal_men.append(country_df[men][silver].Medal.count())
        total_bronze_medal_men.append(country_df[men][bronze].Medal.count())
        total_gold_medal_women.append(country_df[women][gold].Medal.count())
        total_silver_medal_women.append(country_df[women][silver].Medal.count())
        total_bronze_medal_women.append(country_df[women][bronze].Medal.count())
        
        print("Country Name: {} \n Total Medal: {}\n Total Gold Medal: {}\n Total Silver Medal: {}\n Total Bronze Medal: {}\n".format(all_country[j],country_df.Medal.count(),country_df[gold].Medal.count(),country_df[silver].Medal.count(),country_df[bronze].Medal.count()) +
              " Total Men Medal Winner: {}\n Total Women Medal Winner: {}\n".format(country_df[men].Medal.count(),country_df[women].Medal.count()))
        print("Men Medals\n-----------")
        print("Total Gold Medal Men: ",country_df[men][gold].Medal.count())
        print("Total Silver Medal Men: ",country_df[men][silver].Medal.count())
        print("Total Bronze Medal Men: ",country_df[men][bronze].Medal.count())
        print("\nWomen Medals\n-----------")
        print("Total Gold Medal Women: ",country_df[women][gold].Medal.count())
        print("Total Silver Medal Women: ",country_df[women][silver].Medal.count())
        print("Total Bronze Medal Women: ",country_df[women][bronze].Medal.count())
        
        print("*************************************************************************************\n")
        
#         men = df[country] == 'Men'
#         women = df[country] == 'Women'
#         print(df.Medal.count())
        
#         print(j)
        

        
        j +=1
    i +=1
```

# Create Data Sets and merge two data sets

```python
dataSet1 = list(zip(all_country,total_gold_medal_men,total_silver_medal_men,total_bronze_medal_men,total_medal_men))
dataSet2 = list(zip(all_country,total_gold_medal_women,total_silver_medal_women,total_bronze_medal_women,total_medal_women))


dataSet = dataSet1 + dataSet2
dataSet
```

# Create Data frame

```python
df1 = panda.DataFrame(data=dataSet,columns=['Country','Gold','Silver','Bronze','Total Medal'])

df1.groupby(['Country','Total Medal']).sum()
```
